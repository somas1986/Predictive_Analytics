# Box Plot for Day of the Week

```python
sns.boxplot(x='day_of_week', y='attend', data=df_dodgers).set(
    xlabel='Day of the Week', 
    ylabel='Attendance'
)
plt.xticks(rotation=45)
plt.show()
```

As per the Box plot above, we can see that the Tuesdays have the maximum attendacne and Mondays have the least.

# Box Plot for Month

```python
sns.boxplot(x='month', y='attend', data=df_dodgers).set(
    xlabel='Month', 
    ylabel='Attendance'
)
plt.xticks(rotation=45)
plt.show()
```

As per the Box plot above, we can see that the month of June have the maximum attendacne and month of October have the least.

# Box Plot for Day Night

```python
sns.boxplot(x='day_night', y='attend', data=df_dodgers).set(
    xlabel='Month', 
    ylabel='Attendance'
)
plt.xticks(rotation=45)
plt.show()
```

As per the Box plot above, we can't find a big difference in attedance either its day or night.
